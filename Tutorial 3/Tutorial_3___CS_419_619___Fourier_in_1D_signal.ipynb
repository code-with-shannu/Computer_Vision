```python
import numpy as np
import matplotlib.pyplot as plt

samples = [1, 2, 2, 1]

dft = np.fft.fft(samples)

print(dft)
```

```python
samples = [1, 2, 4, 8, 4, 2, 1]

dft = np.fft.fft(samples)

for item in dft:
  print(item)
```

```python
# Shift the zero-frequency component to the center of the spectrum.
# This operation is useful for visualization and analysis.
# It aligns the low frequencies at the center of the spectrum.

dft_shift = np.fft.fftshift(dft)

for item in dft_shift:
  print(item)
```

```python
import matplotlib.pyplot as plt
import numpy as np

freq_bins = np.arange(len(samples))
magnitude_dft = np.abs(dft)
magnitude_dft_shift = np.abs(dft_shift)

shifted_freq_bins = np.arange(-len(samples)//2 + 1, len(samples)//2 + 1, 1)

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.stem(freq_bins, magnitude_dft, 'b', basefmt=" ")
plt.title('Magnitude Spectrum of DFT')
plt.xlabel('Frequency Bin')
plt.ylabel('Magnitude')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.stem(shifted_freq_bins, magnitude_dft_shift, 'r', basefmt=" ")
plt.title('Magnitude Spectrum of Shifted DFT')
plt.xlabel('Frequency Bin')
plt.ylabel('Magnitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
idft_result_shifted = np.fft.ifft(np.fft.ifftshift(dft_shift))

print(idft_result_shifted)
```

```python
idft_result = np.fft.ifft(dft)

print(idft_result)
```
